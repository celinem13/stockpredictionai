# Stocks 4 Update

This is the last week for the stocks project and things started getting better. Animal Crossing stocks were back to high selling point with a buying price of 110 Bells, which is the highest buying price for turnips in the game. It was a major turning point for me financially in the game in order for me to progress and continue stock research. The in-real-life stock market also started to rise in prices, so I was glad I held onto my Google stocks and sold at the highest price I could. The prices started to dip again this afternoon so it's possible Animal Crossing stock prices will start decreasing as well. It was interesting overall to see how similar Animal Crossing: New Horizons' mini-game stock market is to the real stock market. It seems to correlate with the S&P 500 stocks with how it fluctuates with the peaks and valleys. However, there are times when it doesn't correlate at all, seeing how Animal Crossing had a tremendous spike this week while the real stock market had a moderately small spike after hard losses the last two weeks.

I would say the experiment is a success with an overall positive revenue from the stock market simulator from Investopedia.com and the stock market from Animal Crossing: New Horizons. I learned much more about the stock market and feel more comfortable investing in real life rather than playing a simulation or a game. It is possible the RNG (random number generator) was on my side and it actually correlated with the stock market. However, there was no feasible way to prove causation between the two. It would be extremely complicated to take in real-world data and implement real stocks data to a child's game, so it would be highly safe to assume it is based on the game's RNG to create some sort of simulation of a simple stocks market. Overall, I had a fun time with this project and feel more confident to trade and invest.
